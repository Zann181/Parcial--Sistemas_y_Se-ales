### Solución Paso a Paso para la Distancia entre Dos Señales Periódicas

Para calcular la distancia entre dos señales periódicas $x_1(t)$ y $x_2(t)$, utilizaremos la siguiente fórmula:
$$
d^2(x_1, x_2) = \lim_{T \to \infty} \frac{1}{T} \int_{0}^{T} |x_1(t) - x_2(t)|^2 \, dt
$$
Dadas las señales:
- $x_1(t) = Ae^{j\omega_0 t}$
- $x_2(t) = Be^{j5\omega_0 t}$

donde $\omega_0 = \frac{2\pi}{T}$, y $T, A, B \in \mathbb{R}^+$.

#### Paso 1: Integras y pasas a binomio cuadrado perfecto
$$
\frac{1}{T} \left[\int_0^T  |x_1(t)|^2 \,dt -2  \int_0^T x_1(t)x_2(t)\,dt + \int_0^T  |x_2(t)|^2 \,dt\right]
$$
#### La ecuacion queda

$$ P_{x_1}  - \frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right] + P_{x_2}$$


#### Paso 2: Calcular $ P_{x_1}$ y $P_{x_2}$

Calculamos $ P_{x_1}$

$$P_{x_1} = \frac{1}{T} \int_{0}^{T} |x_1(t)|^2 \, dt$$

$$\frac{1}{T} \int_{0}^{T} |Ae^{j\omega_0 t}|^2 \, dt = \frac{A^2}{T} \int_{0}^{T} e^{j\omega_0 t} \cdot e^{-j\omega_0 t} \, dt$$

$$\frac{A^2}{T} \int_{0}^{T} e^{0} \, dt = \frac{A^2}{T} \int_{0}^{T} 1 \, dt = \frac{A^2}{T} [t]_{0}^{T}$$

$$\frac{A^2}{T} [T-0] = A^2$$

$$P_{x_1} = A^2$$

Calculamos $ P_{x_2}$

$$P_{x_2} = \frac{1}{T} \int_{0}^{T} |x_2(t)|^2 \, dt$$

$$\frac{1}{T} \int_{0}^{T} |Be^{5j\omega_0 t}|^2 \, dt = \frac{B^2}{T} \int_{0}^{T} e^{10j\omega_0 t} \cdot e^{-10j\omega_0 t} \, dt$$

$$\frac{B^2}{T} \int_{0}^{T} e^{0} \, dt = \frac{B^2}{T} \int_{0}^{T} 1 \, dt = \frac{B^2}{T} [t]_{0}^{T}$$

$$\frac{B^2}{T} [T-0] = b^2$$

$$P_{x_2} = B^2$$

#### Paso 3: Encontramos el valor de $\frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right]$

$$
\frac{2}{T} \int x_1(t) x_2(t) dt = \frac{2}{T} \int (Ae^{j\omega_0 t})(Be^{j5\omega_0 t}) dt = \frac{2AB}{T} \int e^{6j\omega_0 t} dt
$$

tenemos que $\omega_0 = \frac{2\pi}{T}$ lo remplazamos en la ecuacion
$$
\frac{2AB}{T} \int e^{-j \frac{12\pi}{T} t} dt
$$
Se aplica un cambio de variable
$$
u = 6 \frac{2\pi}{T} t
$$
$$
du = \frac{12j\pi }{T} dt
$$
$$
dt = \frac{T}{12j\pi} du
$$
la ecuacion queda
$$
   \frac{2AB}{T} \int e^{u} \frac{T}{12j\pi} du
$$



####Resolvemos la ecuacion


$$
\frac{2AB}{T} \frac{T}{12\pi} \int_0^T e^u du
$$
##Tenemos que
$$
e^{\frac{12Tj\pi}{T}}= 1
$$
$$
e^{0}= 1
$$
$$
\frac{AB}{6\pi} \left[ e^{\frac{12Tj\pi}{T}} - e^{0}\right]
$$

$$\frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right]
= \frac{AB}{6\pi} \left[ 1- 1 \right]= 0
$$


#### Paso 4: Remplazamos los valores en la ecuacion para encontrar la formula de la distecnia entre  $ P_{x_1}$ y $P_{x_2}$
$$ P_{x_1}  - 0 + P_{x_2}$$

Valores remplazados

$$
P_{x_1-x_2} = A^2 +B^2
$$


#### Conclusión

La distancia media cuadrada entre las señales $x_1(t)$ y $x_2(t)$ es la suma de los cuadrados de las amplitudes $A$ y $B$, lo cual muestra que la distancia no depende de las frecuencias de las señales ni de la variable de tiempo $t$.


```python

from sympy import *

# Definimos los símbolos
t, T, A, B, w0 = symbols('t T A B w0', real=True, positive=True)

# Definimos las señales
x1 = A * exp(I * 2*pi/T* t)
x2 = B * exp(I * 5 * 2*pi/T * t)

# Calculamos la diferencia entre las señales
diff_signal = abs(x1 - x2)

# Calculamos la potencia media de la diferencia
P_mean_diff =  1/T * integrate(diff_signal**2, (t, 0, T))

# Imprimimos el resultado
print('La potencia media de la diferencia entre las señales es:')
P_mean_diff.simplify()
```

### Solución Paso a Paso para la Distancia entre Dos Señales Periódicas

Para calcular la distancia entre dos señales periódicas $x_1(t)$ y $x_2(t)$, utilizaremos la siguiente fórmula:
$$
d^2(x_1, x_2) = \lim_{T \to \infty} \frac{1}{T} \int_{0}^{T} |x_1(t) - x_2(t)|^2 \, dt
$$
Dadas las señales:
- $x_1(t) = Ae^{j\omega_0 t}$
- $x_2(t) = Be^{j5\omega_0 t}$

donde $\omega_0 = \frac{2\pi}{T}$, y $T, A, B \in \mathbb{R}^+$.

#### Paso 1: Expandir el cuadrado y simplificar
$$
\frac{1}{T} \left[\int_0^T  |x_1(t)|^2 \,dt - 2  \int_0^T x_1(t)x_2(t)\,dt + \int_0^T  |x_2(t)|^2 \,dt\right]
$$
La ecuación se simplifica a:
$$
P_{x_1}  - \frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right] + P_{x_2}
$$

#### Paso 2: Calcular $P_{x_1}$ y $P_{x_2}$

Calculamos $P_{x_1}$:
$$
P_{x_1} = \frac{1}{T} \int_{0}^{T} |x_1(t)|^2 \, dt
$$
$$
\frac{1}{T} \int_{0}^{T} |Ae^{j\omega_0 t}|^2 \, dt = \frac{A^2}{T} \int_{0}^{T} e^{j\omega_0 t} \cdot e^{-j\omega_0 t} \, dt
$$
$$
\frac{A^2}{T} \int_{0}^{T} e^{0} \, dt = \frac{A^2}{T} \int_{0}^{T} 1 \, dt = \frac{A^2}{T} [t]_{0}^{T}
$$
$$
\frac{A^2}{T} [T-0] = A^2
$$
$$
P_{x_1} = A^2
$$

Calculamos $P_{x_2}$:
$$
P_{x_2} = \frac{1}{T} \int_{0}^{T} |x_2(t)|^2 \, dt
$$
$$
\frac{1}{T} \int_{0}^{T} |Be^{5j\omega_0 t}|^2 \, dt = \frac{B^2}{T} \int_{0}^{T} e^{10j\omega_0 t} \cdot e^{-10j\omega_0 t} \, dt
$$
$$
\frac{B^2}{T} \int_{0}^{T} e^{0} \, dt = \frac{B^2}{T} \int_{0}^{T} 1 \, dt = \frac{B^2}{T} [t]_{0}^{T}
$$
$$
\frac{B^2}{T} [T-0] = B^2
$$
$$
P_{x_2} = B^2
$$

#### Paso 3: Calcular el término cruzado
$$
\frac{2}{T} \int_0^T x_1(t) x_2(t) dt = \frac{2AB}{T} \int_0^T e^{j\omega_0 t} e^{j5\omega_0 t} dt = \frac{2AB}{T} \int_0^T e^{6j\omega_0 t} dt
$$

Sustituyendo $\omega_0 = \frac{2\pi}{T}$:
$$
\frac{2AB}{T} \int_0^T e^{j\frac{12\pi}{T}t} dt
$$

Hacemos el cambio de variable:
$$
u = 6 \frac{2\pi}{T} t \quad \text{y} \quad du = 6 \frac{2\pi}{T} dt
$$

Entonces:
$$
dt = \frac{T}{12\pi} du
$$

Sustituyendo en la integral:
$$
\frac{2AB}{T} \int e^{u} \cdot \frac{T}{12\pi j} du = \frac{AB}{6\pi j} \int e^{u} du
$$

Resolvemos la integral:
$$
\int e^{u} du = e^{u} = e^{6j \frac{2\pi}{T} t}
$$

Volviendo a la variable original:
$$
\frac{AB}{6\pi j} \cdot e^{6j \frac{2\pi}{T} t} \bigg|_0^T = \frac{AB}{6\pi j} \left( e^{6j \cdot 2\pi} - e^{0} \right)
$$

Como $e^{6j \cdot 2\pi} = 1$:
$$
\frac{AB}{6\pi j} \left( 1 - 1 \right) = 0
$$

Por lo tanto, el término cruzado es cero:
$$
\frac{2}{T} \int_0^T x_1(t) x_2(t) dt = 0
$$

#### Paso 4: Encontrar la distancia entre las señales
Finalmente, la distancia entre las dos señales es:
$$
d^2(x_1, x_2) = A^2 + B^2
$$

Por lo tanto, la distancia entre las señales $x_1(t)$ y $x_2(t)$ es:
$$
d(x_1, x_2) = \sqrt{A^2 + B^2}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la señal continua x(t)
def x(t):
    return 3*np.cos(1000*np.pi*t) + 5*np.sin(2000*np.pi*t) + 10*np.cos(11000*np.pi*t)

# Parámetros de muestreo
Fs = 5000  # Frecuencia de muestreo en Hz
Ts = 1/Fs   # Periodo de muestreo en segundos
t_max = 0.005  # Duración de la señal en segundos para visualización
t = np.arange(0, t_max, Ts)  # Vector de tiempo para la señal discreta

# Generar señal continua y discreta
t_cont = np.linspace(0, t_max, 1000)  # Más puntos para una curva suave en la señal continua
x_cont = x(t_cont)
x_disc = x(t)

# Graficar señal continua
plt.figure(figsize=(12, 6))
plt.plot(t_cont, x_cont, label='Señal continua $x(t)$')

# Graficar señal discreta
plt.stem(t, x_disc, 'r', markerfmt='ro', basefmt=" ", linefmt='r', label='Señal discreta muestreada a 5 kHz', use_line_collection=True)

plt.title('Comparación de la señal continua y discreta')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

## Punto C:

Implemente una simulación para encontrar la salida del sistema lineal e invariante al tiempo $\mathcal{H}\{\cdot\}$, con respuesta al escalón $h_\epsilon[n]=\{2,4,1,5,0,10\}$, $n\in\mathbb{Z}$, ante la entrada análoga $x(t)=20(\cos(t/3)+\cos(t/4))$. Incluya los acondicionamientos necesarios de
discretizacion y cuantización, asumiendo un microprocesador de 4 bits con entrada analoga de 4mA a 20mA.

## Solución:

Revisando la señal $x(t)$, se tiene que:

$\frac{\omega_1}{\omega_2} =\frac{1/3}{1/4} = \frac{4}{3} \in \mathbb{Q}$; por ende $x(t)$ es cuasiperiódica, con período:

$T_1 = \frac{2\pi}{\omega_1} = \frac{2\pi}{1/3} = 6\pi$

$T_2 = \frac{2\pi}{\omega_2} = \frac{2\pi}{1/4} = 8\pi$

$T = T_1 l = T_2 k$

$T = 6\pi l = 8\pi l$

con $k,l\in \mathbb{Z}.$ El $MCM(6,8)=24,$ así:

$\boxed{T = 24\pi [s]}$

Además, se requiere aplicar cero pendiente:

$\hat{x}(t)=m x(t)+ b$, con:

$m = \frac{\hat{x}_{max}(t)-\hat{x}_{min}(t)}{x_{max}(t)-x_{min}(t)}$

$b = \hat{x}_{min}(t)-m x_{min}(t)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

# función cero y pendiente actualizada para el rango de 4mA a 20mA
def my_ceropend(x, xhmax=20, xhmin=4):
    # cálculo de pendiente y cero
    m = (xhmax - xhmin) / (x.max() - x.min())
    b = xhmin - m * x.min()
    xh = m * x + b
    return xh

# Parámetros de la señal y simulación actualizados
T1 = 6 * np.pi  # periodo señal 1
T2 = 8 * np.pi  # periodo señal 2
F1 = 1 / T1
F2 = 1 / T2
T = 24 * np.pi  # periodo cuasiperiódica
Fs = 10 * max(F1, F2)  # frecuencia de muestreo actualizada
t = np.arange(0, 3 * T, 1 / Fs)  # vector de tiempo, se muestran 3 periodos
x = 20 * (np.cos(t / 3) + np.cos(t / 4))

# Aplicación de cero y pendiente actualizada para el rango de 4mA a 20mA
xhmax = 0.020  # Máximo  para 20mA
xhmin = 0.004   # Mínimo para 4mA
xh = my_ceropend(x, xhmax, xhmin).reshape(-1, 1)  # aplicar cero y pendiente

# Cuantización para un microprocesador de 4 bits
nbits = 4  # número de bits actualizado
Ne = 2 ** nbits  # cantidad de estados actualizada
res = (xhmax - xhmin) / Ne  # resolución de cuantización
ve = np.arange(xhmin, xhmax, res).reshape(-1, 1)  # vector de estados

xq = np.zeros(len(xh))  # señal cuantizada
for i in range(len(xh)):
    xq[i] = ve[np.argmin(cdist(xh[i].reshape(-1, 1), ve))]

# Respuesta al escalón del sistema actualizada
he = np.array([2, 4, 1, 5, 0, 10])
h = np.diff(he)  # respuesta al impulso del sistema

# Convolver la señal de entrada cuantizada con la respuesta al impulso
y = np.convolve(xq, h, mode='same')  # respuesta del sistema


# Preparar gráficas
fig, axs = plt.subplots(3, 1, figsize=(10, 16))

# Señal original x(t)
axs[0].plot(t, x, 'r', label='$x(t)$')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('Amplitud')
axs[0].set_title('Señal Original $x(t)$')
axs[0].legend()

# Señal acondicionada $\hat{x}(t)$
axs[1].plot(t, xh, 'b', label='$\hat{x}(t)$')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Señal Acondicionada $\hat{x}(t)$')
axs[1].legend()

# Señal cuantizada $x_q[n]$
axs[1].stem(t, xq, 'y', linefmt='y-', markerfmt='yo', basefmt=" ", label='$x_{q}[n]$')
axs[1].set_xlabel('n [muestra]')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Señal Cuantizada $x_q[n]$')
axs[1].legend()

# Respuesta del sistema y[n]
axs[2].stem(t, y, 'b', linefmt='b-', markerfmt='bo', basefmt=" ", label='$y[n]$')
axs[2].set_xlabel('n [muestra]')
axs[2].set_ylabel('Amplitud')
axs[2].set_title('Respuesta del Sistema $y[n]$')
axs[2].legend()

plt.tight_layout()
plt.show()
```
